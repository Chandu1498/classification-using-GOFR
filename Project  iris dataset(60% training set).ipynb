```python
class FDT:
    
    def __init__(self, data):

        self.left = None
        self.right = None
        self.data = data
    
    def construct(self,depth):
        fs=['AND','OR']
        ts=['sl','sw','pl','pw']  
        print("                  ",self.data,'\n')
        self=self.left
        print('                    ',self.data)
        for ct in range(depth-1,0,-1):
            if(ct==1):
                #print("countaaa :",ct)
                x=random.randint(0,3)
                y=random.randint(0,3)
                self.left=FDT(ts[x])
                #print("Left child of",self.data,"is ",self.left.data)
                self.right=FDT(ts[y])
                #print("Right child of",self.data,"is ",self.right.data)
                print(" "*ct,self.left.data,"  "*ct,self.right.data)
            elif(self.data in fs):
                if(random.randint(1,2)==1):
                    #print("counta :",ct)
                    a=random.randint(0,1)
                    self.left=FDT(fs[a])
                    #print("Left child of",self.data,"is ",self.left.data)
                    b=random.randint(0,3)
                    self.right=FDT(ts[b])
                    #print("Right child of",self.data,"is ",self.right.data)
                    print("  "*ct,self.left.data,"  "*ct,self.right.data)
                    self=self.left
                else:
                    #print("count :",ct)
                    a=random.randint(0,1)
                    self.right=FDT(fs[a])
                    #print("Right child of",self.data,"is ",self.right.data)
                    b=random.randint(0,3)
                    self.left=FDT(ts[b])
                    #print("Left child of",self.data,"is ",self.left.data)
                    print("  "*ct,self.left.data,"   "*ct,self.right.data)
                    self=self.right 
    
    def PrintTree(self):
        print(self.data)
        if self.left:
            self.left.PrintTree()
        if self.right:
            self.right.PrintTree()
    
    def PrintTree1(self):
        if self.left:
            print("(", end =" ")
            self.left.PrintTree1()
        print(self.data, end =" ")
        if self.right:   
            self.right.PrintTree1()
            print(")", end =" ")
    
    def bfs(self):
        queue = Queue()
        queue.put(self)
        while not queue.empty():
            current_node = queue.get()
            print(current_node.data,end=" ")

            if current_node.left:
                queue.put(current_node.left)

            if current_node.right:
                queue.put(current_node.right)
            
    def insertData(self,i):
        global sla,swa,pla,pwa,x_sl,x_sw,x_pl,x_pw,y
        ##print("qwerty",self.data)
        if self.data == "sl": 
            ##print("Value :",self.data,sla[i])
            self.data=round(np.interp(sla[i],x_sl,y),2)
            ##print("Triangular value :",self.data)
        if self.data == "sw": 
            ##print("Value :",self.data,swa[i])
            self.data=round(np.interp(swa[i],x_sw,y),2)
            ##print("Triangular value :",self.data)
        if self.data == "pl": 
            ##print("Value :",self.data,pla[i])
            self.data=round(np.interp(pla[i],x_pl,y),2)
            ##print("Triangular value :",self.data)
        if self.data == "pw": 
            ##print("Value :",self.data,pwa[i])
            self.data=round(np.interp(pwa[i],x_pw,y),2)
            ##print("Triangular value :",self.data)
        if self.left:
            self.left.insertData(i)
        if self.right:   
            self.right.insertData(i)
            
def copyFDT(root):
    if root is None:
        return
    ptr=FDT(root.data)
    ptr.left=copyFDT(root.left)
    ptr.right=copyFDT(root.right)
    return ptr
 
def traverse(rootnode):
  thislevel = [rootnode]
  c=15
  while thislevel:
    nextlevel = list()
    for n in thislevel:
      print(" "*c,n.data,end=" ")
      if n.left: nextlevel.append(n.left)
      if n.right: nextlevel.append(n.right)
    print()
    thislevel = nextlevel
    c=c-3
            
def evaluateExpressionTree(root): 
  
    # empty tree 
    if root is None: 
        return 0
  
    # leaf node 
    if root.left is None and root.right is None: 
        return float(root.data) 
  
    # evaluate left tree 
    left_sum = evaluateExpressionTree(root.left) 
  
    # evaluate right tree 
    right_sum = evaluateExpressionTree(root.right) 
  
    # check which operation to apply 
    if root.data == 'AND': 
        return min(left_sum,right_sum) 
    elif root.data == 'OR': 
        return max(left_sum,right_sum)
    
```

```python
def Crossover(tree1,tree2,rand):
    array=[0 for x in range(2)]
    global c1
    print(c1,rand,tree1.data,tree2.data)
    if(c1==rand):
        if(tree1.left.left):
            subTree1=copyFDT(tree1.left)
        else:
            subTree1=copyFDT(tree1.right)
        traverse(subTree1)
        if(tree2.left.left):
            subTree2=copyFDT(tree2.left)
        else:
            subTree2=copyFDT(tree2.right)
        traverse(subTree2)
        if(tree1.left.left):
            tree1.left=subTree2
        elif(tree1.right.left):
            tree1.right=subTree2
        if(tree2.left.left):
            tree2.left=subTree1
        elif(tree2.right.left):
            tree2.right=subTree1
    if(tree1.left.left and tree2.left.left):
        c1=c1+1
        Crossover(tree1.left,tree2.left,rand)
    if(tree1.right.left and tree2.right.left):
        c1=c1+1
        Crossover(tree1.right,tree2.right,rand)
    if(tree1.left.left and tree2.right.left):
        c1=c1+1
        Crossover(tree1.left,tree2.right,rand)
    if(tree1.right.left and tree2.left.left):
        c1=c1+1
        Crossover(tree1.right,tree2.left,rand)
    array[0]=copyFDT(tree1)
    array[1]=copyFDT(tree2)
    return array
```
